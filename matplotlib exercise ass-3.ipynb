```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## exercise 1

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("title")
```

## exercise 2

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])
```

```python
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

fig
```

## exercise 3

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
```

```python
ax.plot(x,z)
ax.set_xlabel('x')
ax.set_ylabel('z')
ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax2.set_title('zoom')
fig
```

## exercise 4

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
axes[0].plot(x,y,color='blue',lw=3,ls="--")
axes[1].plot(x,z,color='red',lw=3,ls='-')
fig
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,2))
axes[0].plot(x,y,color='blue',lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].plot(x,z,color='red',lw=3,ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
